### Bagging: Round 1

```python
#Bootstrapped sample 1
boot = iris.sample(len(iris), replace=True) 
#boot.head()
```

```python
uniq_indx = np.unique(boot.index)
test_indx = []

for i in range(0,len(iris)):
    if i not in uniq_indx:
        test_indx.append(i)
        
#print(uniq_indx)
#print(test_indx)
```

```python
#Out-of-bag data
test = iris.iloc[test_indx]
#test.head()
```

```python
#Splitting X and Y
X_boot = boot.drop("Species", axis=1)
Y_boot = boot[["Species"]]

X_test = test.drop("Species", axis=1)
Y_test = test[["Species"]]


#print(X_boot.head())
#print(Y_boot.head())
```

```python
#Training the model on the bootstrap sample
tree = DecisionTreeClassifier(criterion = "gini", min_samples_leaf=5)
tree.fit(X_boot, Y_boot)
```

```python
#OOB-prediction
y_pred = tree.predict(X_test)
#y_pred
```

### Bagging in General

```python
#Creating a separate data frame to save the OOB prediction
pred = {}


#Number of bagging rounds
nrounds = 20


#data
data = iris


for k in range(0,nrounds):
    #Draw a bootstrap sample
    boot = iris.sample(len(iris), replace=True) 


    #To get the OOD data
    uniq_indx = np.unique(boot.index)
    test_indx = []
    for i in range(0,len(iris)):
        if i not in uniq_indx:
            test_indx.append(i)


    #Out-of-bag data
    test = iris.iloc[test_indx]


    #Splitting X and Y
    X_boot = boot.drop(["Species"], axis=1)
    Y_boot = boot[["Species"]]

    X_test = test.drop(["Species"], axis=1)
    Y_test = test[["Species"]]


    #Training the model on the bootstrap sample
    tree = DecisionTreeClassifier(criterion = "gini", min_samples_leaf=5)

    tree.fit(X_boot, Y_boot)


    #OOB-prediction
    y_pred = tree.predict(X_test)


    #Attaching this OOB prediction to our data set
    pred1 = np.repeat(np.nan,len(iris))
    pred1 = np.array(pred1, dtype = '<U10')
    pred1[test_indx] = y_pred
    key = "Pred" + str(k+1)
    pred[key] = pred1

```

```python
pred = pd.DataFrame(pred)
pred
```

```python
#For each rows find the frequency of each classes

def fin_predicted(x):
    d = {}
    for i in x:
        d[i] = d.get(i,0) + 1
    
    #Remove the nan key
    d.pop('nan', None)
    key = list(d.keys())
    val = list(d.values())
    return(key[val.index(max(val))])

```

```python
#Testing the CODE 1
row1 = pred.iloc[0,]
fin_predicted(row1)
```

```python
#Testing the CODE 2
row3 = pred.iloc[3,]
fin_predicted(row3)
```

```python
#Obtaining the final predicted values for each obs based on the  majority classes
fin_pred = []

for indx in range(0,len(pred)):
    fin_pred.append(fin_predicted(pred.iloc[indx,]))
    
fin_pred[:10]
```

```python
#Attaching it to our data
iris2 = iris.copy()
iris2["Predicted"] = fin_pred
iris2
```

```python
c2 = confusion_matrix(iris2.Species, iris2.Predicted)
c2
```

## Bagging for Classification Problem

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix
```

```python
iris=pd.read_csv('iris.csv')
iris.describe()
```

```python
data=iris.drop('Id',axis=1)
data.head()
```

```python
pred=pd.DataFrame()
pred["Pred1"] = np.array(np.repeat(np.nan,len(iris)))
pred.head()
```

### Bagging Round 1

```python
boot = data.sample(len(data),replace=True)
uniq_idx = np.unique(boot.index)
test_idx=[]
for i in data.index:
    if i not in uniq_idx:
        test_idx.append(i)
X_train=data.iloc[uniq_idx].drop('Species',axis=1)
y_train=data.iloc[uniq_idx]['Species']
X_test=data.iloc[test_idx].drop('Species',axis=1)
y_test=data.iloc[test_idx]['Species']
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
#Decision Tree
dt=DecisionTreeClassifier(criterion='gini',min_samples_leaf=5,max_depth=4)
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
```

```python
pred.loc[test_idx,'Pred1']=np.array(y_pred)
pred[1:10]
```

```python
for j in range(2,20):
    uniq_idx=[]   #unique index 
    boot=[]       #bootstrap sample
#     print(str(j))
    pred["Pred"+str(j)]=np.array(np.repeat(np.nan,len(data)))  #new column
    boot = data.sample(len(data),replace=True)  #creating bootstrap samples
    uniq_idx = np.unique(boot.index)   #unique values
    test_idx=[]    
    for i in data.index:
        if i not in uniq_idx:
            test_idx.append(i)   #getting test data
    X_train=data.iloc[uniq_idx].drop('Species',axis=1)   #Train X
    y_train=data.iloc[uniq_idx]['Species']              #Train y
    X_test=data.iloc[test_idx].drop('Species',axis=1)  #Test X
    y_test=data.iloc[test_idx]['Species']              #test y
#Decision Tree
    dt=DecisionTreeClassifier(criterion='gini',min_samples_leaf=5,max_depth=4)  #dt classifier
    dt.fit(X_train,y_train)
    y_pred=dt.predict(X_test)    #predicting values
    pred.loc[test_idx,'Pred'+str(j)]=np.array(y_pred)   #adding columns
#     print(j)
pred
```

```python
final_pred=np.array(pred.mode(axis=1))   #majority values
```

```python
confusion_matrix(data['Species'],final_pred)  #confusion matrix
```

```python
from sklearn.metrics import accuracy_score    #accuracy score
auc=accuracy_score(data['Species'],final_pred)
auc
```
